```python
# Imported libraries used

import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Importing csv for manipulations

spy_2008 = pd.read_csv(
    Path('SPY_2008.csv'), index_col='Date'
)

spdr_2008 = pd.read_csv(
    Path("Consumer Staples Select Sector SPDR Fund.csv"), 
    index_col='Date'
)

vanguard_2008 = pd.read_csv(
    Path('Vanguard Consumer Staples ETF .csv'), 
    index_col='Date'
)

spy_2015 = pd.read_csv(
    Path('Flat Market 2015 SPY.csv'), 
    index_col='Date'
)

spdr_2015 = pd.read_csv(
    Path('Flat Market 2015 Consumer Staples Select Sector SPDR Fund.csv'), 
    index_col='Date'
)

vanguard_2015 = pd.read_csv(
    Path('Flat Market 2015 Vanguard Consumer Staples ETF .csv'), 
    index_col='Date'
)

spy_2018 = pd.read_csv(
    Path('Politcal Period 2018 SPY.csv'),
    index_col='Date'
)

spdr_2018 = pd.read_csv(
    Path('Politcal Period 2018 Consumer Staples Select Sector SPDR Fund.csv'),
    index_col='Date'
)

vanguard_2018 = pd.read_csv(
    Path('Politcal Period 2018 Vanguard Consumer Staples ETF  (1).csv'),
    index_col='Date'
)

spy_10 = pd.read_csv(
    Path('10 Year Overall Outlook - SPY.csv'),
    index_col='Date'
)

spdr_10 = pd.read_csv(
    Path('10 Year Overall Outlook - Consumer Staples Select Sector SPDR Fund.csv'),
    index_col='Date'
)

vanguard_10 = pd.read_csv(
    Path('10 Year Overall Outlook - Vanguard Consumer Staples ETF .csv'),
    index_col='Date'
)
```

```python
# Data concatination for all dataframes

combined_data_2008 = pd.concat(
    [spy_2008, spdr_2008, vanguard_2008], axis="columns", join="inner" 
                        )
combined_data_2008.sort_index(inplace=True)
combined_data_2008.columns = ['SPY', 'SPDR', 'Vanguard']

combined_data_2015 = pd.concat(
    [spy_2015, spdr_2015, vanguard_2015], axis="columns", join="inner"
)
combined_data_2015.sort_index(inplace=True)
combined_data_2015.columns = ['SPY', 'SPDR', 'Vanguard']


combined_data_2018 = pd.concat(
    [spy_2018, spdr_2018, vanguard_2018], axis="columns", join="inner"
)
combined_data_2018.sort_index(inplace=True)
combined_data_2018.columns = ['SPY', 'SPDR', 'Vanguard']

combined_data_10 = pd.concat(
    [spy_10, spdr_10, vanguard_10], axis="columns", join="inner"
)
combined_data_10.sort_index(inplace=True)
combined_data_10.columns = ['SPY', 'SPDR', 'Vanguard']
```

```python
# Functions used to gather daily returns, volatility, and cumulative returns.

def daily_returns(x):
        daily_returns = x.pct_change().dropna()
        return daily_returns

def volatility(x):
    volatility = (daily_returns(x).std() * np.sqrt(252))
    vol = float(volatility)
    percent_volatility = '{:.2%}'.format(vol)
    print(percent_volatility)
        
def cumulative_returns(x):
    cumulative_returns = (1 + daily_returns(x)).cumprod().dropna()
    return cumulative_returns
```
