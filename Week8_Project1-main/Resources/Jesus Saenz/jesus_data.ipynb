```python
# Functions used to gather daily returns, volatility, and cumulative returns.

def returns(x):
    ## Calculate daily returns of closing prices for 
    ## ETFs, Stock Tickers, or Indices.
    daily_returns = x.pct_change().dropna()
    
    def sharpe(daily_returns):
        ## The ratio is the average return earned in 
        ## excess of the risk-free rate per unit of volatility or total risk. 
        return np.sqrt(252) * (daily_returns.mean() / daily_returns.std())
    
    def cumulative_returns(daily_returns):
        ## Calculate the cumulative returns of the 
        ## ETFs, Stock Tickers, or Indices.
        cumulative_returns = (1 + daily_returns(x)).cumprod().dropna()
        return cumulative_returns
    
    print(f'''
    The daily returns of the sectors are:
    {daily_returns}
    
    The cumulative returns of the sectors are:
    {cumulative_returns}
    
    The average returns in excess of risk-free
    rate per unit of volatility:
    {sharpe(daily_returns)}
    ''')

def volatility(x):
    ## Returns the percent volatility of the 
    ## ETFs, Stock Tickers, or Indices.
    volatility = (daily_returns(x).std() * np.sqrt(252))
    vol = float(volatility)
    percent_volatility = '{:.2%}'.format(vol)
    print(f'The volatility is: {percent_volatility}')
```

```python
portfolio_2008 = pd.concat(
    [spdr_2008_js, vanguard_2008_js, xlu_2008_js],
    axis="columns",
    join="inner"
                        )

portfolio_2015 = pd.concat(
    [spdr_2015_js, vanguard_2015_js, xlu_2015_js],
     axis="columns",
    join="inner"
                        )

portfolio_2018 = pd.concat(
    [spdr_2018_js, vanguard_2018_js, xlu_2018_js],
     axis="columns",
    join="inner"
                        )

portfolio_10_years = pd.concat(
    [spdr_10_js, vanguard_10_js, xlu_10_js],
     axis="columns",
    join="inner"
)
```

```python
def daily_returns(x):
    ## Calculate daily returns of closing prices for 
    ## ETFs, Stock Tickers, or Indices.
    daily_returns = x.pct_change().dropna()
    
    return daily_returns
```

```python
weights = [0.3, 0.4, 0.3]

weighted_port_2008 = daily_returns(portfolio_2008).dot(weights)

weighted_port_2015 = daily_returns(portfolio_2015).dot(weights)

weighted_port_2018 = daily_returns(portfolio_2018).dot(weights)

weighted_port_10 = daily_returns(portfolio_10_years).dot(weights)
```
